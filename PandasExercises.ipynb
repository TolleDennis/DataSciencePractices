Aggregate Statistics (Groupby)

```python
#grouping data
dennis=pd.read_csv("pokemon_data.csv")
dennis.groupby(["Type 1"]).mean().sort_values("HP",ascending=False)
```

```python
dennis.groupby(["Type 1"]).mean().sort_values("HP",ascending=True)
```
